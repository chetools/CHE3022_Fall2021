```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default='plotly_dark'
from scipy.optimize import newton_krylov
```

```python
nr, ntheta = 25, 27
dr, dtheta = 1./(nr-1), 7/4*np.pi/(ntheta-1)
r, theta = np.mgrid[1:4:(nr*1j), 0:np.pi/2:(ntheta*1j)]

T_in, T_out = 130, 150
T_plo, T_phi = 100, 150


def residual(T):
    d2r = np.zeros_like(T)
    d2theta = np.zeros_like(T)
    
    d2r[1:-1,:] = ((r[1:-1,:]+dr/2) *(T[2:,:]-T[1:-1,:]) - (r[1:-1,:]-dr/2)*(T[1:-1,:]-T[:-2,:]))/dr/dr 
    d2r[0,:]    =((r[0,:]+dr/2)*(T[1,:]-T[0,:]) - (r[0,:]-dr/2)*(T[1,:]-T_in)) /dr/dr 
    d2r[-1]   = ((r[-1,:]+dr/2)*(T_out -T[-1,:]) -(r[-1,:]-dr/2)*(T[-1,:]-T[-2,:]))/dr/dr 

    d2theta[:,1:-1] = (T[:,2:] - 2*T[:,1:-1] + T[:,:-2])/dtheta/dtheta/r[:,1:-1]
    d2theta[:,0]    = (T[:,1]  - 2*T[:,0]    + T_plo)/dtheta/dtheta/r[:,0]
    d2theta[:,-1]   = (T_phi   - 2*T[:,-1]   + T[:,-2])/dtheta/dtheta/r[:,-1]

    return d2r + d2theta

# solve
guess = np.zeros((nr, ntheta))
sol = newton_krylov(residual, guess, method='lgmres', verbose=0)
```

```python

x,y = r*np.cos(theta), r*np.sin(theta)

fig = make_subplots()
for i in np.arange(nr):
     fig.add_trace(go.Scatter3d(x=x[i,:], y=y[i,:], z=sol[i,:], mode='lines', line_color='gray'))

for j in np.arange(ntheta):
     fig.add_trace(go.Scatter3d(x=x[:,j], y=y[:,j], z=sol[:,j], mode='lines', line_color='gray'))

fig.add_trace(go.Surface(x=x,y=y,z=sol))

fig.update_layout(width=800, height=800, scene=dict(xaxis_title='x', yaxis_title='y', zaxis_title='T'),
                  showlegend=False)
```
